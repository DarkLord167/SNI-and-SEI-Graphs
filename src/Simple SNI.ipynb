```python
import numpy as np
import networkx as nx
```

```python
def set_nodes_degree(graph, delta_prime, node):
    global node_num
    node_num = node
    for i in range(delta_prime):
        x = delta_prime
        node_num += 1
        graph.add_node(node_num)
        graph.add_edge(node, node_num)
        if (x >= 0):
            x = x - 1
            set_nodes_degree(graph, x, node_num)
```

```python
def cons_sni_graph(delta):
    sni_graph = nx.Graph(graph_type='SNI')
    sni_graph.add_node('p')
    if(delta < 2):
        print("Can not construct SNI with graph degree of" , delta)
    else:
        global node_num
        node_num = 0
        for i in range(delta):
            node_num += 1
            sni_graph.add_node(node_num)
            sni_graph.add_edge('p', node_num)
            delta_prime = delta - 2
            set_nodes_degree(sni_graph, delta_prime, node_num)
    return sni_graph




sni_graph = cons_sni_graph(4)
nx.draw_networkx(sni_graph, node_color='yellow')
```

```python
print(len(sni_graph.nodes))
```

```python
# Get adjacency matrix
A = nx.to_numpy_array(sni_graph)
print(A)
```

```python
# Calculate det of adjacency matrix
print(A.shape)
print(np.linalg.det(A))
```
